<strong><span class="color7"> C4 - Biosignalsnotebooks Functions</span></strong>

<strong><span>C4.1 - Importing relevant packages</span></strong>

```python
# Biosignalsnotebooks python package
import biosignalsnotebooks as bsnb
```

<strong><span>C4.2 Load data from signal samples library</span></strong>

```python
# Load of data
relative_file_path = "../../signal_samples/emg_1000_hz_16_bits_solo.txt"
data, header = bsnb.load(relative_file_path, get_header=True)
```

<strong><span>C4.3 Install the needed package</span></strong>

Run this code only once, to install the needed package:

<img src="../../images/other/quick_start_guide/img15.png" style="display:block;height:100%,width:100%">

<strong><span>C4.4 Check File Header Data and making your code more colorful</span></strong>

```python
from sty import fg, rs
print(fg(98,195,238) + "\033[1mHeader\033[0m" + fg.rs)
print(header)
print(fg(98,195,238) + "\033[1mData Structure\033[0m" + fg.rs)
print(data)
```

<strong><span>C4.5 Store information from file Header</span></strong>

```python
#get information which is stored inside variables
ch = "CH1" # Channel
sr = header["sampling rate"] # Sampling rate
resolution = header["resolution"][0] # Resolution (number of available bits)
device = header["device"]
```

<strong><span>C.4.6 - Store the desired data in an individual variable</span></strong>

```python
#RAW DATA
signal = data[ch]
```

<strong><span>C.4.7 - Convert the RAW data to values with a physical meaning (in this case electric voltage | mV)</span></strong>

```python
# Signal Samples Conversion
signal_uv = bsnb.raw_to_phy("EMG", device, signal, resolution, option="mV") # Conversion to mV
```

<strong><span>C.4.8 - Generate a time-axis</span></strong>

```python
time = bsnb.generate_time(signal)
```

<strong><span>C.4.9 - PLot Data</span></strong>

```python
bsnb.plot([time], [signal_uv], y_axis_label="Electric Tension (mV)")
```
